# 1. Background

Drowning is one of the leading causes of death in children 0-4 years of age. Data from the Australian Bureau of Statistics shows that drowning is the number one cause of death in one-, two- and three-year-old children (ABS 2019). Between 1 July 2002/03 and 30 June 2018/19, 496 children aged 0-4 years drowned in Australia. Of these, 202 (41%) children aged one year drowned – within the context of an entire lifespan, from 0 to 100 years, no other age is at greater risk of drowning.

Most deaths in children aged one occurred in backyard swimming pools (59%) and during the summer months (44%). Accidental falls into water was the leading activity prior to drowning in this age group, accounting for 88% of all deaths. Almost all of these deaths were due to a lack of active adult supervision. For every toddler drowning death approximately eight children are admitted to hospital as a result of non-fatal drowning. 

Royal Life Saving through their program, ***The Keep Watch program***, suggest four key drowning prevention actions: Supervise, Restrict Access, Water Awareness and Resuscitate. When implemented together, these safety measures can help to maximise child safety around water.

The core conviction of the above program is however, that the child is the most important part of a drowning prevention strategy, i.e., Water Awareness. Children are curious, capable, and have an uncanny ability to overcome obstacles like pool fences; Therefore preparing them water survival skills to potentially save themselves if they find themselves in the water alone, is indeed essential.

The main objective of this report is to provide best estimation on where to open infant survival swim school based on the data from Australian Biro Statistic (ABS) website.

### 2.1 Age Demographic Distribution Data

The excel file name is ***demography_2018.xls***, contains data about the distribution of resident in Australia. Before Data being process and analysed, Data is explored by looking into the inside the Excel file to gain better understanding on what data is presented and how they are presented.

After careful exploration, It is understood that Data is separated by 3 categories, male, female, and persons and stored in three excel sheets, namely `Table 1`, `Table 2` and `Table 3`.

As it really doesn't matter which gender of our potential baby customer, `Table 3` is selected for further process.

```python
# Import data as dfraw

dfraw=pd.read_excel('demography_2018.xls', sheet_name = 'Table 3')
print('Data Size is {}'.format(dfraw.shape))
dfraw.head(15)
```

As seen from above Dataframe, data preprocess is required so that data are meaningful to be used. Meaningful Data are located in columns:
* `Unnamed: 1` - showing which state the data belongs to
* `Unnamed: 9` to `Unnamed: 28` - Showing the distribution population of Age Group

Few steps needs to be done:
* Remove row 0 - 4, 6 and 7
* Remove any unused columns for example: `Australian Bureau of Statistics`, `Unnamed: 2`, to `Unnamed: 8`.
* Remove any other `NaN` on the rows

```python
# Remove row 
dfraw.drop([0,1,2,3,4,6,7], inplace = True)

# Remove any unnecessary column
dfraw.drop(['Unnamed: 2', 'Unnamed: 3','Unnamed: 4','Unnamed: 5','Unnamed: 6',
            'Unnamed: 7','Unnamed: 8'], axis=1, inplace = True)

dfraw.drop(dfraw.columns[[0]], axis=1, inplace = True)
dfraw.reset_index(drop = True, inplace = True)
dfraw.head(10)
dfraw.drop(dfraw.iloc[:, 12:], axis=1, inplace = True)
dfraw
```

Now the data is look more readable and representative. Next we are going to rename the columns, with States, Suburb and Age Group (from `index` 0).

Note that check with original `Excel` file is also being done to ensure that no important information is lost during data cleaning process.



```python
# Take the age group from index 0 and then drop index 0
age = dfraw.iloc[0,:]
age.dropna(inplace = True)
age
dfraw.drop([0], inplace = True)

clm=['State', 'Suburb']
for i,ag in enumerate(age):
    clm.append(age[i])

print('Compare dataframe columns size and clm list size')
print('DataFrame size : {} vs Column List Size: {}'. format(len(dfraw.columns), len(clm)))    


dfraw.columns = clm
dfraw.tail(15)
```

There are still `NaN` on the Data, we will remove the NaN that is consistent through out whole row

```python
indexNames = dfraw[dfraw['State'].isnull()].index
print(indexNames)

dfraw.drop(indexNames , inplace=True, axis=0)
dfraw.reset_index(drop = True, inplace = True)
print(dfraw.shape)
dfraw
```

```python
#Now extract WA data
indexWA = dfraw[dfraw['State'] != "Western Australia"].index

indexWA
dfraw.drop(indexWA , inplace=True, axis=0)
dfraw.reset_index(drop = True, inplace = True)
dfraw
```

### 2.2 Income Data

The excel file name is ***income_2019.xls***, contains data of personal income distribution in Australia. Before Data being process and analysed, Data is explored by looking into the inside the Excel file to gain better understanding on what data is presented and how they are presented.
