```python
import pandas as pd
import folium
from folium import plugins
```

```python
#read data
path = r'C:\\Users\\admin\\Documents\\Kintha\\Python\\Capstone\\Tutorialjson\\'
df = pd.read_csv('{}'.format(path) + r"Final_data.csv")
```

```python
#Convert Infant to string for plot
df['Infant'] = df['Infant'].astype('str')

# create a plain Perth Metro Area
LGA_map = folium.Map(location=[-31.8093, 115.9468], zoom_start=10)

# instantiate a mark cluster object for the incidents in the dataframe - No. of infant
infants = plugins.MarkerCluster().add_to(LGA_map)

# loop through the dataframe and add each data point to the mark cluster
for lat, lng, label1 in zip(df.Latitude, df.Longitude, df.Infant):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label1,
#        icon=folium.Icon(color='blue',icon='bar-chart', prefix='fa') 
    ).add_to(infants)

# instantiate a mark cluster object for the incidents in the dataframe
lga = plugins.MarkerCluster().add_to(LGA_map)  

# display map
LGA_map
```

```python
LGA_map = folium.Map(location=[-31.8093, 115.9468], zoom_start=10)
LGA_geo = '{}'.format(path) + 'LGA_WA.sjon'

df['Infant'] = df['Infant'].astype('float64')

# generate choropleth map using the Number of Infant
LGA_map.choropleth(
     geo_data=LGA_geo,
     data=df,
     columns=['LGA','Infant'],
     key_on='features.properties.wa_lga_s_3',
     fill_color='YlOrRd', 
     fill_opacity=0.7, 
     line_opacity=0.2,
     legend_name='Number of Infant'
    )

# display map
LGA_map
```
