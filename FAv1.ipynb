# 1. Background

Drowning is one of the leading causes of death in children 0-4 years of age. Data from the Australian Bureau of Statistics shows that drowning is the number one cause of death in one-, two- and three-year-old children (ABS 2019). Between 1 July 2002/03 and 30 June 2018/19, 496 children aged 0-4 years drowned in Australia. Of these, 202 (41%) children aged one year drowned – within the context of an entire lifespan, from 0 to 100 years, no other age is at greater risk of drowning.

Most deaths in children aged one occurred in backyard swimming pools (59%) and during the summer months (44%). Accidental falls into water was the leading activity prior to drowning in this age group, accounting for 88% of all deaths. Almost all of these deaths were due to a lack of active adult supervision. For every toddler drowning death approximately eight children are admitted to hospital as a result of non-fatal drowning. 

Royal Life Saving through their program, ***The Keep Watch program***, suggest four key drowning prevention actions: Supervise, Restrict Access, Water Awareness and Resuscitate. When implemented together, these safety measures can help to maximise child safety around water.

The core conviction of the above program is however, that the child is the most important part of a drowning prevention strategy, i.e., Water Awareness. Children are curious, capable, and have an uncanny ability to overcome obstacles like pool fences; Therefore preparing them water survival skills to potentially save themselves if they find themselves in the water alone, is indeed essential.

The main objective of this report is to provide best estimation on where to open infant survival swim school based on the data from Australian Biro Statistic (ABS) website.

### 2.2 Income Data

The excel file name is ***income_lga_2019.xls***, contains data of personal income distribution in Australia. Before Data being process and analysed, Data is explored by looking into the inside the Excel file to gain better understanding on what data is presented and how they are presented.

After careful exploration, It is understood that Data is separated by 5 total income categories: by Greater Capital City, Statistical area level 2 to 4 and by Local Government Area. These data stored in 5 excel sheets, namely `Table sf.1`, `Table 2f.2`, `Table 2f.3`, `Table 2f.4` and `Table 2f.5`

What we are interested in is data by Local Government Area, `Table 2f.5` and it is selected for further process.

```python
# Import data as dfraw

dfinc_raw=pd.read_excel('income_lga_2019.xls', sheet_name = 'Table 2f.5')
print('Data Size is {}'.format(dfinc_raw.shape))
dfinc_raw.head(10)
```

As seen from above Dataframe, data preprocess is required so that data are meaningful to be used. Meaningful Data are located in columns:
* `Unnamed: 1` - showing which suburb the data belongs to
* `Unnamed: 6` - Showing Mean income for each Suburb

Few steps needs to be done:
* Remove row 0 - 3
* Remove any unused columns for example: `Australian Bureau of Statistics`, `Unnamed: 2`, to `Unnamed: 8`.
* Remove any other `NaN` on the rows

```python
dfinc_raw.dropna(inplace = True)
dfinc_raw.drop(dfinc_raw.iloc[:,7:], axis = 1, inplace = True)
dfinc_raw.drop(dfinc_raw.iloc[:,2:6], axis = 1, inplace = True)

#Creating column Names
dfinc_raw.reset_index(drop = True, inplace = True)
col = ['State Code', 'LGA','Income']
dfinc_raw.columns = col
dfinc_raw.head()

#indexing WA data
indexwa = []
i=0
for i in range(0,len(dfinc_raw)):  
    if str(dfinc_raw.iloc[i,0]).startswith('5') == True:
        indexwa.append(i)
    i=+1
dfinc_WA = dfinc_raw.iloc[indexwa[0]:indexwa[len(indexwa)-1]+1,:].reset_index(drop = True)
dfinc_WA.drop('State Code', axis = 1, inplace = True)
print('Size of the dataframe {}'.format(dfinc_WA.shape))
dfinc_WA.head()
```

### 2.3 List of Suburb

```python
url = 'https://en.wikipedia.org/wiki/List_of_Perth_suburbs'
dfsuburb_raw = pd.read_html(url)
dfsuburb = dfsuburb_raw[2]

# Split LGA data separated by '; '
dfsuburb.columns = ['Suburb', 'LGA']
print(dfsuburb.shape)
check2 = dfsuburb[dfsuburb['LGA'].str.contains("; ")].index
check2

# Create temp DF for storing double or triple LGA
dfraw = dfsuburb.iloc[check2,:]
#split LGA
temp = dfraw.iloc[:,1].str.split('; ')
dfraw['temp'] = temp
dfraw.drop('LGA', axis = 1, inplace = True)

#Explode LGA
dfraw = dfraw.explode('temp')
dfraw.columns = ['Suburb', 'LGA']
dfsuburb
dfraw.head()

#original DF
dfsuburb.drop(check2, inplace = True)
dfsuburb.head()

# concat original DF and explode DF
dfsb = pd.concat([dfsuburb, dfraw], ignore_index=False, sort =False, axis=0).reset_index(drop=True)
dfsb.head()
# Group Suburb Based on their LGA
dfsbs = dfsb.groupby('LGA', sort = False).agg(', '.join).reset_index(drop = False)
dfsbs.head()
```

### 2.4 Geospatial Data

The Geospatial data consist of Suburb in Perth, WA and its Geographical position coordinates

```python
# Import Table
url2='https://www.geonames.org/postal-codes/AU/WA/western-australia.html'
dfs = pd.read_html(url2)
dfgeo_raw = pd.DataFrame(data=dfs[2])
dfgeo_raw.drop(['Unnamed: 0', 'Country', 'Admin1','Admin2','Admin3'], axis=1, inplace=True)
dfgeo_raw.dropna(axis=0,inplace=True)
dfgeo_raw.head()
dfgeo_raw.columns = ['Suburb', 'Postal Code']
dfgeo_raw.head()
```

```python
# Indexing geographical position in table
endIn = len(dfgeo_raw)
indexll = pd.Series(range(1,endIn,2))
long, lat=[],[]
for i in indexll:
    long.append(dfgeo_raw.iloc[i,0].split('/')[0])
    lat.append(dfgeo_raw.iloc[i,1].split('/')[1])
longlat = pd.DataFrame({'Longitude': long, 'Latitude': lat})
longlat

#remove geographical location from table
dfgeo_raw.drop(index=indexll,inplace=True)
dfgeo_raw.reset_index(drop=True, inplace = True)

dfgeo_raw.head()
print(dfgeo_raw.shape)
dfgeo = dfgeo_raw.join(longlat)
dfgeo.head()
```

### 2.5 Merge Four data

```python
#Demography
dfdemo_WA
#Income
dfinc_WA
# Merge Demographic and Income Data
df_demoinc = pd.merge(dfdemo_WA, dfinc_WA, how = 'inner', on='LGA')

# Merge Suburb and Geospatial Data
df1 = pd.merge(df_demoinc,dfsbs, how = 'inner', on = 'LGA')

# Merged all data together as final data
dfgeo
dfgeo.columns = ['LGA','Postal Code', 'Longitude', 'Latitude']
df = pd.merge(df1,dfgeo,how='inner', on='LGA')
```

```python
#Re-arrange columns
df = df[['Postal Code', 'LGA', 'Suburb', '0-4','Parents','Income', 'Longitude', 'Latitude']]

#Change Data Type
df['0-4'] = df['0-4'].astype('float64')
df['Income'] = df['Income'].astype('float64')
df['Longitude'] = df['Longitude'].astype('float64')
df['Latitude'] = df['Latitude'].astype('float64')

#Final Dataframe
print('Dataframe Size: {}'.format(df1.shape))
df
```
